```python
import json
```

```python
with open('csvs/google_topics.json', 'r+', encoding='utf-8') as f:
    google_topic = json.loads(f.read())

with open('csvs/gdelt_topics.json', 'r+', encoding='utf-8') as f:
    gdelt_topic = json.loads(f.read())
```

```python
google_topic_set = [set(i) for i in google_topic]
gdelt_topic_set = [set(i) for i in gdelt_topic]
```

```python
intersection_topics = [set.intersection(google_topic_set[i], gdelt_topic_set[i]) for i in range(len(google_topic))]
```

```python
cross_intersection_topics = []
for i in google_topic_set:
    cross_intersection_topics.append([set.intersection(i, gdelt_topic_set[j]) for j in range(len(gdelt_topic_set))])
```

```python
connections = []
weights = []

for i, google in enumerate(cross_intersection_topics):
    for j, gdelt in enumerate(cross_intersection_topics[i]):
        
        if len(gdelt) > 0:
            connections.append((i+10, j+30))
            weights.append(len(gdelt))
```

```python
import json
import numpy as np

import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_graph(graph, weight, node_labels=None, edge_labels=None, graph_layout='shell',
               node_size=1600, node_color='blue', node_alpha=0.3,
               node_text_size=12,
               edge_color='blue', edge_alpha=0.3, edge_tickness=1,
               edge_text_pos=0.3,
               text_font='sans-serif'):

    G=nx.Graph()

    for edge, w in zip(graph, weight):
        G.add_edge(edge[0], edge[1], weight=w)

    
    if graph_layout == 'spring':
        graph_pos=nx.spring_layout(G)
    elif graph_layout == 'spectral':
        graph_pos=nx.spectral_layout(G)
    elif graph_layout == 'random':
        graph_pos=nx.random_layout(G)
    else:
        graph_pos=nx.shell_layout(G)

    if edge_labels is None:
        edge_labels = range(len(graph))

    edge_labels = dict(zip(graph, edge_labels))
    
    labels_ = {}    
    if node_labels is not None:
        for node in G.nodes():
            labels_[node] = node_labels[node]
    
    
    nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, label=node_labels,
                           alpha=node_alpha, node_color=node_color)
    nx.draw_networkx_edges(G,graph_pos,width=edge_tickness,
                           alpha=edge_alpha,edge_color=edge_color)
    nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,
                            font_family=text_font)

    
    nx.draw_networkx_labels(G,graph_pos,labels_,font_size=16,font_color='r')
#     try:
#         nx.draw_networkx_edge_labels(G, graph_pos, edge_labels=edge_labels, 
#                                      label_pos=edge_text_pos)
#     except:
#         pass

    plt.show()
```

```python
plt.figure(figsize=(20,10))
graph, weight = connections, np.array(weights) / max(weights)

topic_node_labels = {**{i: f"Google {i-10}" for i in range(10,20)}, **{i: f"GDELT {i-30}" for i in range(30,40)}}
draw_graph(graph, weight, node_labels=topic_node_labels)
```

```python
with open('Graph_data/cross/graph_values.json', 'w+', encoding='utf-8') as f:
    f.write(json.dumps(graph))
    
with open('Graph_data/cross/graph_weights.json', 'w+', encoding='utf-8') as f:
    f.write(json.dumps(weights))
```
