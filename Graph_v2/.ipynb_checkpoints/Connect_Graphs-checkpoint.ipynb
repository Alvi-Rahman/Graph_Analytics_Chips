```python
def draw_graph(graph, weight=None, node_labels=None, edge_labels=None, graph_layout='spring',
               node_size=1600, node_color='blue', node_alpha=0.3,
               node_text_size=12,
               edge_color='blue', edge_alpha=0.3, edge_tickness=1,
               edge_text_pos=0.3,
               text_font='sans-serif'):

    G=nx.Graph()

    for edge in graph:
        G.add_edge(edge[0], edge[1])

    
    if graph_layout == 'spring':
        graph_pos=nx.spring_layout(G)
    elif graph_layout == 'spectral':
        graph_pos=nx.spectral_layout(G)
    elif graph_layout == 'random':
        graph_pos=nx.random_layout(G)
    else:
        graph_pos=nx.shell_layout(G)

#     if edge_labels is None:
#         edge_labels = range(len(graph))

#     edge_labels = dict(zip(graph, edge_labels))
    
    labels_ = {}    
    if node_labels is not None:
        for node in G.nodes():
            labels_[node] = node_labels[node]
    
    
    nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, label=node_labels,
                           alpha=node_alpha, node_color=node_color)
    nx.draw_networkx_edges(G,graph_pos,width=edge_tickness,
                           alpha=edge_alpha,edge_color=edge_color)
    nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,
                            font_family=text_font)

    
    nx.draw_networkx_labels(G,graph_pos,labels_,font_size=16,font_color='r')
#     try:
#         nx.draw_networkx_edge_labels(G, graph_pos, edge_labels=edge_labels, 
#                                      label_pos=edge_text_pos)
#     except:
#         pass

    plt.show()
```

```python
plt.figure(figsize=(20,10))

topic_node_labels = {**{i: f"Google {i-10}" for i in range(10,20)}, **{i: f"GDELT {i-30}" for i in range(30,40)}}
google_company_node_labels = {i: list(thirty_sorted_freq_dict_google.keys())[i] for i in range(len(thirty_sorted_freq_dict_google.keys()))}
gdelt_company_node_labels = {i+20: list(thirty_sorted_freq_dict_gdlet.keys())[i] for i in range(len(thirty_sorted_freq_dict_gdlet.keys()))}

all_node_labels = {**topic_node_labels, **google_company_node_labels, **gdelt_company_node_labels}
graph = all_connections

draw_graph(graph, node_labels=all_node_labels)
```

```python
haha=nx.DiGraph()
for edge in graph:
    haha.add_edge(edge[0], edge[1])

strongs = sorted(nx.strongly_connected_components(haha), key=len, reverse=True)

for strong in strongs[:2]:
    plt.figure(figsize=(20, 10))
    draw_graph(haha.subgraph(strong).edges(), node_labels=all_node_labels)
```

```python
def draw_deep_graph(graph, weight=None, node_labels=None, edge_labels=None, graph_layout='spring',
               node_size=1600, node_color='blue', node_alpha=0.3,
               node_text_size=12,
               edge_color='blue', edge_alpha=0.3, edge_tickness=1,
               edge_text_pos=0.3,
               text_font='sans-serif'):

    G=nx.Graph()

    for edge in graph:
        G.add_edge(edge[0], edge[1])

    g = dgl.DGLGraph(G)
    
    G = g.to_networkx()
    
    if graph_layout == 'spring':
        graph_pos=nx.spring_layout(G)
    elif graph_layout == 'spectral':
        graph_pos=nx.spectral_layout(G)
    elif graph_layout == 'random':
        graph_pos=nx.random_layout(G)
    else:
        graph_pos=nx.shell_layout(G)

#     if edge_labels is None:
#         edge_labels = range(len(graph))

#     edge_labels = dict(zip(graph, edge_labels))
    
    labels_ = {}    
    if node_labels is not None:
        for node in G.nodes():
            labels_[node] = node_labels[node]
    
    
    nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, label=node_labels,
                           alpha=node_alpha, node_color=node_color)
    nx.draw_networkx_edges(G,graph_pos,width=edge_tickness,
                           alpha=edge_alpha,edge_color=edge_color)
    nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,
                            font_family=text_font)

    
    nx.draw_networkx_labels(G,graph_pos,labels_,font_size=16,font_color='r')
#     nx.draw_networkx_edge_labels(G, graph_pos, edge_labels=edge_labels, 
#                                  label_pos=edge_text_pos)

    plt.show()
    
    return g, G
```

```python
plt.figure(figsize=(20,10))

small_g, large_G = draw_deep_graph(graph, node_labels=all_node_labels)
```

```python
N = len(small_g.nodes())  # number of nodes
DAMP = 0.85  # damping factor
K = 10  # number of iterations

small_g.ndata['pv'] = torch.ones(N) / N
small_g.ndata['deg'] = small_g.out_degrees(small_g.nodes()).float()
```

```python
def pagerank_message_func(edges):
    return {'pv' : edges.src['pv'] / edges.src['deg']}
```

```python
def pagerank_reduce_func(nodes):
    msgs = torch.sum(nodes.mailbox['pv'], dim=1)
    pv = (1 - DAMP) / N + DAMP * msgs
    return {'pv' : pv}
```

```python
import dgl.function as fn

def pagerank_func(g):
    g.ndata['pv'] = g.ndata['pv'] / g.ndata['deg']
    g.update_all(message_func=fn.copy_src(src='pv', out='m'),
                 reduce_func=fn.sum(msg='m',out='m_sum'))
    g.ndata['pv'] = (1 - DAMP) / N + DAMP * g.ndata['m_sum']
```

```python
for k in range(K):
    # Uncomment the corresponding line to select different version.
    # pagerank_naive(g)
    # pagerank_batch(g)
    # pagerank_level2(g)
    pagerank_func(small_g)
print(small_g.ndata['pv'])
```
